```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
train = pd.read_csv("data/train.csv")
raw = pd.read_csv("data/ObesityDataSet_raw_and_data_sinthetic.csv")
```

```python
# train 데이터의 범주형 변수를 숫자형으로 인코딩
label_encoders = {}
for column in train.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    train[column] = le.fit_transform(train[column])
    label_encoders[column] = le
```

```python
# train 데이터의 독립 변수와 종속 변수 분리
train_X = train.drop('NObeyesdad', axis=1)
train_Y = train['NObeyesdad']
```

```python
# SMOTE 적용
smote_tr = SMOTE(random_state=42)
X_resampled, y_resampled = smote_tr.fit_resample(train_X, train_Y)
```

```python
# SMOTE 적용 후 클래스 분포 확인
print(y_resampled.value_counts())
```

```python
# 균형 잡힌 데이터셋을 생성한 후의 데이터프레임을 만듭니다.
balanced_data = pd.concat([X_resampled, y_resampled], axis=1)
balanced_data
```

```python
# raw 데이터의 범주형 변수를 숫자형으로 인코딩
label_encoders = {}
for column in raw.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    raw[column] = le.fit_transform(raw[column])
    label_encoders[column] = le
```

```python
# 독립 변수와 종속 변수 분리
raw_X = raw.drop('NObeyesdad', axis=1)
raw_Y = raw['NObeyesdad']
```

```python
# SMOTE 적용
smote_raw = SMOTE(random_state=42)
X_resampled_raw, y_resampled_raw = smote_raw.fit_resample(raw_X, raw_Y)
```

```python
# SMOTE 적용 후 클래스 분포 확인
print(y_resampled_raw.value_counts())
```

```python
# 균형 잡힌 데이터셋을 생성한 후의 데이터프레임을 만듭니다.
balanced_data_raw = pd.concat([X_resampled_raw, y_resampled_raw], axis=1)
balanced_data_raw
```

```python
data = pd.concat([balanced_data, balanced_data_raw], axis=0)
data
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.pop("id")
```

```python
data["NObeyesdad"].value_counts()
```

```python
# 독립 변수와 종속 변수 분리
data_X = data.drop('NObeyesdad', axis=1)
data_Y = data['NObeyesdad']
```

```python
# 예시: 각 클래스별 추가할 샘플 수 계산
additional_samples_per_class = 2746  # 각 클래스당 추가할 샘플 수
target_samples_per_class = {i: 4397 + additional_samples_per_class for i in range(7)}  # 각 클래스의 목표 샘플 수

# SMOTE 적용
smote = SMOTE(sampling_strategy=target_samples_per_class, random_state=42)
X_resampled, y_resampled = smote.fit_resample(data_X, data_Y)

# 결과 확인
print('새로운 데이터셋의 크기:', X_resampled.shape)
```

```python
data = pd.concat([X_resampled, y_resampled], axis=1)
data
```

```python
data.to_csv("data/data.csv")
```
